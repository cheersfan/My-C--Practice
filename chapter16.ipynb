**基于范围的for循环(C++11）**  
基于范围的for循环和java里集合类(例如ArrayList）的for循环很像，却别就是，C++的可以修改变量的值，java不行。  
C++中修改容器内容需要一个引用参数，如下:  
```C++
//方法的参数为引用参数
void InflateReview（Review & r){r.rating++}；
for（auto &x : books){
    InflateReview（x)；
}
```
Java中的使用方法类似:  
```Java
ArrayList<Integer> mylist = new ArrayList<Integer>;
mylist.add（1)；
mylist.add（2)；
mylist.add（3)；
for（int x ： mylist){
   System.out.print（x + " “);
}

```
