# 第8章-函数探幽  

## 8.5函数模板  
对函数模板有些陌生，随着C++发展对模板的`具体化`的也有新的发展，记录如下。  

* 模板的定义 
* 模板的显式具体化、具体化、实例化  

---

### 模板的定义  
```C++
//声明
template <typename T>
void myswap(T &a, T &b);

//定义
template <typename T>
void myswap(T &a , T &b){
    T temp;
    temp = a;
    a = b;
    b = temp;
}
```

这里`template <typename>`在声明和定义中都要在模板函数的前面。仅声明时出现，会报错。
**注** ： 在模板函数前声明两个typename会报错，所以模板函数的泛型定义只能声明成一中泛型类型吗？模板函数是否可以声明多个参数类型不一样？

### 模板的显式具体化  
一个函数原型可以有多种形式：  
> 1.非模板函数
> 2.模板函数  
> 3.显式具体化模板函数  
> 4.重载函数  

编译器在选择函数原型时的优先级为：非模板函数 > 显式具体化模板函数 > 模板函数  

显式具体化模板函数的定义，需要以`template<>`开头，并通过名称来指出类型。例子如下：
```C++
//job是一个已定义的结构体P285
template <> void myswap<job>(job&, job&);
```
以上例子的`<job>`可以省略。这个`<job>`的含义是，标明这个模板函数的泛型将具体化为`int`型。


### 模板的实例化和具体化  
在实际代码中，函数模板本身并不会生成函数的定义，只有在具体使用函数模板时，才会根据参数类型生成相应的函数定义，得到的是模板的`实例（instantiation）`。函数模板的实例化可分为：  
> 隐式实例化  
> 显式实例化（explicit instantiation）  

`隐式实例化`是默认的，当编译器根据参数类型将模板函数生成相应的函数定义。  
`显式实例化`需要程序员显式的声明。具体示例如下，将myswap()模板函数显式示例化为int型：
```C++
template void myswap<int>(int, int);
```

`显式实例化`与模板的具体化的区别：前者用template声明，后者用template<>声明，且后者可以将函数名后面的`<>`省略。  

**具体化（specialization）**包含：
> 1. 隐式实例化
> 2. 显式实例化
> 3. 显式具体化
    
三者的相同之处在于表示的都是使用具体类型的函数定义，而不是通用的描述（比如函数模板）。
