**函数签名**：主要与函数参数的数目、类型和参数的排列顺序有关。两个同名函数有一个不一样都可以成为是重载函数，反之则不是。

### 模板的定义  
```C++
//声明
template <typename T>
void myswap(T &a, T &b);

//定义，在具体的Clion IDE中，这里的template<typename T>肯能会被自动简化为<T>
template <typename T>
void myswap(T &a , T &b){
    T temp;
    temp = a;
    a = b;
    b = temp;
}
```

这里`template <typename>`在声明和定义中都要在模板函数的前面。仅声明时出现，会报错。

### 模板的显式具体化  
一个函数原型可以有多种形式：  
> 1.非模板函数  
> 2.模板函数  
> 3.显式具体化模板函数  
> 4.重载函数  

**显示具体化(explicit specialization)** 编写的函数模板不能处理某些类型，例如swap()函数交换两个值，但若想交换两个结构体中的一个值时，定义的模板函数则不支持。此时，需要定一个具体化函数，并**定义**。

编译器在选择函数原型时的优先级为：非模板函数 > 显式具体化模板函数 > 模板函数  

显式具体化模板函数的定义，需要以`template<>`开头，并通过名称来指出类型。例子如下：
```C++
//job是一个已定义的结构体P285
template <> void myswap<job>(job&, job&);
```
以上例子的`<job>`可以省略。这个`<job>`的含义是，表明这个模板函数的泛型将具体化为`job`型。


### 模板的实例化和具体化  
在实际代码中，函数模板本身并不会生成函数的定义，只有在具体使用函数模板时，才会根据参数类型生成相应的函数定义，得到的是模板的`实例（instantiation）`。函数模板的实例化可分为：  
> 隐式实例化  
> 显式实例化（explicit instantiation）  

`隐式实例化`是默认的，当编译器根据参数类型将模板函数生成相应的函数定义。  
`显式实例化`需要程序员显式的声明函数的参数类型，不需要编译器根据调用时函数的参数来推测函数的参数类型。具体示例如下，将myswap()模板函数显式示例化为int型：
```C++
template void myswap<int>(int, int);
```



**具体化（specialization）**包含：
> 1. 隐式实例化
> 2. 显式实例化
> 3. 显式具体化
    
`隐式实例化`是编译器默认进行的，根据调用模板函数的函数参数来自动推测参数类型。    
`显式实例化`与模板的`显式具体化`的区别：在定义方面，前者用template声明，后者用template<>声明，且后者可以将函数名后面的`<>`省略。在编译方面，由于模板函数只有在具体调用时才会根据模板的定义生成对应参数的函数，显式实例化只是声明了要使用的参数类型，不需要编译器根据调用时数据类型进行推测，而显式具体化则是重新定义了函数体，会被编译器优于模板函数考虑。
